```python
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
import numpy as np
import csv
import glob

def Read_Data(WeekNumber, Year):
      columns=["Mode", "Date", "DateTime", "CardID", "CardType", "VehicleID",
                   "ParentRoute", "RouteID", "StopID"] 
      fname = glob.glob(Year + '/Week%s' %WeekNumber + "/*.txt")
      print(fname)
      dataset = pd.read_csv(fname[0], sep="|", names=columns)
      return dataset

    
def Append_data(Year, FirstWeek, LastWeek):
    basic_table = Read_Data(FirstWeek, Year)
    for i in range(FirstWeek+1, LastWeek+1):
          Actual_Table    = Read_Data(i, Year)
          basic_table     = basic_table.append(Actual_Table) 
          print('appended', len(basic_table))
    return basic_table   

# we only read in weeks 2 - 25 for all datasets, so they are consistent     
dataset_2016 = Append_data("2016",2,25)#25
#dataset_2017 = Append_data("2017",2,25)
dataset_2018 = Append_data("2018",2,25)
```

```python
def group(mydata):  
    # how many touch-ons belong to each CardID
    mydata_touchOns = mydata.groupby("CardID").DateTime.count().reset_index()
    return mydata_touchOns
    
dataset_2016_grouped = group(dataset_2016)
dataset_2017_grouped = group(dataset_2017)
dataset_2018_grouped = group(dataset_2018)
```

```python
plt.figure(figsize=(8,6))
ax = plt.subplot()

plt.hist(dataset_2016_grouped.DateTime, bins=30, label='2016', color = 'blue', normed = True, histtype = 'step', linewidth=2)
plt.hist(dataset_2017_grouped.DateTime, bins=30, label='2017', color = 'green', normed = True, histtype = 'step', linewidth=2)
plt.hist(dataset_2018_grouped.DateTime, bins=30, label='2018', color = 'grey', normed = True, histtype = 'step', linewidth=2)

plt.title('Myki usage 2016-2018 (Jan-June)',fontsize=18)
plt.xlabel('Nr. of touch-ons',fontsize=18)
plt.legend()
#plt.yscale('log')
plt.ylabel('Nr. of customers',fontsize=18)
plt.savefig("MykiUsage_2016_2018.png")
plt.show()
```
